To use this cleaning function and to receive an impression a function is used. The clean_and_plot function is with the total_count function able to count on unique values and plot the result ordered.

```python
from collections import defaultdict
def total_count(df, col1, col2, look_for):
    '''
    INPUT:
    df - the pandas dataframe you want to search
    col1 - the column name you want to look through
    col2 - the column you want to count values from
    look_for - a list of strings you want to search for in each row of df[col]

    OUTPUT:
    new_df - a dataframe of each look_for with the count of how often it shows up
    '''
    new_df = defaultdict(int)
    #loop through list of ed types
    for val in look_for:
        #loop through rows
        for idx in range(df.shape[0]):
            #if the ed type is in the row add 1
            if val in df[col1][idx]:
                new_df[val] += int(df[col2][idx])
    new_df = pd.DataFrame(pd.Series(new_df)).reset_index()
    new_df.columns = [col1, col2]
    new_df.sort_values(col2, ascending=False, inplace=True)
    return new_df
```

```python
def clean_and_plot(df, column, title='Plot', plot=True, column_name='count'):
    '''
    INPUT 
        df - a dataframe with the values to count
        colum - what colum to inspect
        title - string the title of your plot
        plot - bool providing whether or not you want a plot back
        
    OUTPUT
        study_df - a dataframe with the count of how many individuals
        Displays a plot of pretty things related to the inspected column.
    '''
    possible_vals = []
    for dbs in df[column]:
        for db in str(dbs).split(';'):
            possible_vals.append(db)
    
    possible_vals = set(possible_vals)
    
    study = df[column].value_counts().reset_index()
    study.rename(columns={'index': 'method', column: column_name}, inplace=True)
    study_df = total_count(study, 'method', column_name, possible_vals)

    study_df.set_index('method', inplace=True)
    if plot:
        (study_df/study_df.sum()).plot(kind='bar', legend=None);
        plt.title(title);
        plt.show()
    props_study_df = study_df/study_df.sum()
    return props_study_df
```

As the best readable plot for a blogpost I would choose the horizontal bar chart. In the next step a function is written to deploy each category in the same pattern.

```python
def compare_and_plot(df, column):
    '''
    INPUT 
        df - a dataframe with the values to count
        colum - what colum to inspect
        
    OUTPUT
        Displays a plot comparing the WorkedWith and DesireNextYear column. (Both must exist)
    '''
    df_19 = clean_and_plot(df, column+'WorkedWith', plot=False, column_name = '2019')
    df_20 = clean_and_plot(df, column+'DesireNextYear', plot=False, column_name = '2020')
    
    df_both = df_19.merge(df_20, on='method')

    a = df_both.plot(kind="barh", figsize=(8, df_both.shape[0]/2.0))
    a.invert_yaxis()
    plt.xlabel("Percent", size=14)
    plt.ylabel(column, size=14)
    a.legend(loc='center right')
    
```
